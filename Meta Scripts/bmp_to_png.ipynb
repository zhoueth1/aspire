```python
from pathlib import Path
from PIL import Image

# Notebook is in: Meta Scripts/
meta_scripts_dir = Path.cwd()

# Sibling folders
project_root = meta_scripts_dir.parent
raw_data_dir = project_root / "Data" / "Raw Data"
pngs_dir = project_root / "Data" / "PNGs"

pngs_dir.mkdir(exist_ok=True)

bmp_files = list(raw_data_dir.rglob("*.bmp"))

total_files = len(bmp_files)

print(f"Found {total_files} BMP files")

if total_files == 0:
  print("No BMP files found.")
else:
  for i, bmp_path in enumerate(bmp_files, start=1):
    relative_path = bmp_path.relative_to(raw_data_dir)

    output_path = pngs_dir / relative_path.with_suffix(".png")
    output_path.parent.mkdir(parents=True, exist_ok=True)

    with Image.open(bmp_path) as img:
      img.save(output_path, "PNG")

    percent_done = 100 * i / total_files

    print(
      f"\rConverted {i}/{total_files} files ({percent_done:.1f}%)",
      end=""
    )

  print("\nDone converting BMP files to PNG.")
```
